# A. Object-Oriented Programming (OOP)

### What is init?

in Python, <b>init</b> is a special method used in classes. It's called the constructor and is automatically executed when a new object of the class is created. 

### Purpose of init. 
- It initializes the object's attributes.
- It allows you to pass in values when creating an object. 

```python
class ClassName: 
    def __init__(self, param1, param2): 
        self.param1 = param1 
        self.param2 = param2 
```

```python
class Resistor: 
    def __init__(self, resistance, power_rating):
        self.resistance = resistance 
        self.power_rating = power_rating

r1 = Resistor(220, 0.25) 
print(r1.resistance) 
print(r1.power_rating)
```

We define a class Resistor to simulate a physical resistor component. The constructor __init__ takes two parameters: resistance (in Ohms) and power rating (in Watts). The describe() method returns a string summary of the component.

```python
# Class Resistor 
import numpy as np
class Resistor: 
    def __init__(self, resistance, power_rating):
        self.resistance = resistance 
        self.power_rating = power_rating
        self.voltage = np.sqrt(resistance * power_rating)

    def describe(self): 
        status = "Resistor: {:.2f}  \nWatts: {:.2f} W \nVoltage: {:.2f} V".format(self.resistance, self.power_rating, self.voltage)

        return status
    def parallel_resistance(self, *params): 
        return sum([i ** -1 for i in params]) ** -1

r1 = Resistor(220, 0.25) 
print(r1.describe())
```

We simulate 100 random sensor readings between 20.0 and 30.0, write them to a file (sensor_log.txt), and compute basic statistics like mean, max, and min values.

```python
#Sensor data logging and stats
import random

sensor_data = [random.uniform(20.0, 30.0) for _ in range(100)]
with open("sensor_log.txt", "w") as f:
    for data in sensor_data:
        f.write(f"{data}\n")

mean_val = sum(sensor_data) / len(sensor_data)
max_val = max(sensor_data)
min_val = min(sensor_data)

print(f"Mean: {mean_val:.2f}, Max: {max_val:.2f}, Min: {min_val:.2f}")
```

# C. Matplotlib for Data Visualization

We generate a sine wave using NumPy and plot it using Matplotlib. This simulates an analog signal like voltage over time. We use np.linspace for the time base and np.sin to simulate a periodic signal.

```python
# Task 3: Sine wave simulation with NumPy
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 500)
frequency = 5
sine_wave = np.sin(2 * np.pi * frequency * t)

plt.plot(t, sine_wave)
plt.title("Sine Wave Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

# Anomaly Detection 
#### Rules:

- Temperature Anomaly: If temp is above 90C.
- Efficiency Anomaly: If efficiency drops below 65%.

### Python Code: Load & Process Data

```python
import pandas as pd

# Load dataset
df = pd.read_excel("machine_performance.xlsx")

# Clean column names (removes hidden spaces)
df.columns = df.columns.str.strip()

# Ensure 'Date' is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Debugging: Print available columns
print("Columns in dataset:", df.columns)
```

```python
def detect_anomaly(row):
    if row.get("Temperature", 0) > 90:  # Handles missing values safely
        return 'Temp Anomaly'
    elif row.get("Efficiency", 100) < 65:
        return 'Efficiency Anomaly'
    else:
        return 'Normal'

# Apply anomaly detection
df['Anomaly'] = df.apply(detect_anomaly, axis=1)

# Debugging: Check anomaly column
print(df[['Machine ID', 'Date', 'Anomaly']].head())
```

### Machine Health Summary 

```python
health_summary = df.groupby('Machine ID')['Status'].value_counts().unstack().fillna(0)
print(health_summary)
```

### Plot Time Series Per Machine

```python
import matplotlib.pyplot as plt

machine_ids = df['Machine ID'].unique()

for machine in machine_ids:
    machine_df = df[df['Machine ID'] == machine]
    
    plt.figure(figsize=(10, 4))
    plt.plot(machine_df['Date'], machine_df['Temperature'], label='Temperature (C)', color='red')
    plt.plot(machine_df['Date'], machine_df['Efficiency'], label='Efficiency (%)', color='blue')
    plt.title(f"Machine {machine} Performance Over Time")
    plt.xlabel("Date")
    plt.ylabel("Value")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

### Heatmap of Status Counts

```python
import seaborn as sns

plt.figure(figsize=(6,4))
sns.heatmap(health_summary, annot=True, cmap="coolwarm", fmt='g')
plt.title("Machine Health Status")
plt.show()
```

### Save Anomalies to Excel 

```python
if "Anomaly" in df.columns:
    anomalies = df[df['Anomaly'] != 'Normal']
    anomalies.to_excel("machine_anomalies_report.xlsx", index=False)
    print("Anomaly report saved successfully!")
else:
    print("Error: 'Anomaly' column not found in dataset.")
```
