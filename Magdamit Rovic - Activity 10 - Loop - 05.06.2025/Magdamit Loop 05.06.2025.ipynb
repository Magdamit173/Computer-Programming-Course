# Python Loop

```python
#1. for loop in Python
## A for loop is used to iterate over a sequence (list, tuple, dictionary, string, or range). 
# Syntax. 
# for item in iterable: 
    # do something 
# Example 1: 

for i in range(5): # stop before 5 
    print(i) 
```

```python
# Example 2: for loop: Print ALL Scores 
scores = [85, 92, 78, 66, 90]

print("All Student Scores:") 
for score in scores: 
    print(score)
```

```python
# Example 3: Print ALL Passing Grades ( >= 75) 
print("Passing Grades:") 
for score in scores: 
    if score >= 75: 
        print(score)
```

```python
#2. break statement in for loop 
## The break statement exits the loop immediately when a condition is met. 
# Example 4: 
for i in range(10): 
    if i == 5: 
        break      # stop before 5 
    print(i)
```

```python
# Example 5: break in for loop: Stop if a Failing Grade is Found 
scores = [85, 92, 78, 49, 90]

print("Checking scores:") 
for score in scores: 
    if score < 50: 
        print(f"Failing score detected: {score}")
        break # stop before 90 
    print(f"Passed {score}") 
```

```python
#3. Continue Statement in for Loop
## The continue statement skips the current iteration and continues with the next. 
# Example 6: 
for i in range(5): 
    if i == 2: 
        continue # skips 2
    print(i)
```

```python
# Example 7:
print("Only showing score belows 90:") 
for score in scores:
    if score >= 90: 
        continue
    print(score)
```

```python
# Example 8: Skip Perfect Scores (continue) 
print("\nGrades (excluding perfect 100):") 
for score in scores: 
    if score == 100: 
        continue
    print(score)
```

```python
# Example 10: Multiplication Table (with break)  i, j, k < --- the 3 iterator legends XD
for i in range(1, 6): #1, 2, 3,4 5, 
    for j in range(1, 6): #1, 2, 3,4 5, 
        if j == 4: 
            break
        print(f"{i} x {j} =  {i*j}", end="\t") 
    print() # New Line
# Effect: Inner Loop breaks at 3. Output only shows up to 3 per row. 
```

```python
# Example 11: Skip Even Numbers in Table (continue) 
for i in range(1, 6): 
    for j in range(1, 6): 
        if j % 2 == 0: # 1,3,5 
            continue 
        print(f"{i} x {j} =  {i*j}", end="\t")
    print() 
# Effect: Inner Loop skips even numbers (2, 4) - prints odd - numbered multiples.
```

```python
# Example 12: Seating Chart Checker 

rows = ["A", "B", "C"]
seats = [1,2,3,4]

# rows.reverse() 
# seats.reverse()

for row in rows: 
    for seat in seats: 
        if row == "B" and seat == 3: 
            print(f"Seat {row}{seat} is blocked") 
            continue
        print(f"Seat  {row}{seat} is blocked")
# Use Case: Practical for tracking seat availability with conditions.
```

```python
#### Activity
import numpy as np

students = ["Anna", "Ben", "Cara"]
subjects = ["Math", "English", "Science"] 

grade_system = ["A - Excellent", "B - Very Good", "C - Good", "D - Needs Improvement", "F - Failed"]
grade_system.reverse()



for student in students: 
    print(f"\nGrades for {student}:")
    passed_count = 0
    all_grades = []
    
    for subject in subjects: 
        while True:
            grade = input(f"Enter {student}'s grade in {subject}: ")

            if type(grade) == str and not grade.isdigit():
                print("Input Must be a Number or Non-Negative Number") 
                continue

            grade = int(grade)
            if grade < 50 or grade > 100 or grade < 0: 
                print("Grade must be in range of (50 to 100) And Non-Negative Number")
                continue  
            elif grade >= 50 and grade <= 100: 

                if grade >= 75:
                    passed_count = passed_count + 1
                
                all_grades.append(grade)
                
                grade_filtered = int(grade / 10) - 5
                if grade_filtered == 5: grade_filtered = 4
                if grade_filtered <= 0: grade_filtered = 0
                    
                print(f"Student: {student} - {subject} : {grade_system[grade_filtered]}") 
                break

    
    print("-" * 40) 
    print(f"Summary for {student}")
    print(f" Passed {passed_count} Out Of {len(all_grades)} Subjects")
    print(f"Average Grade: {np.average(all_grades):.2f}")
    print("-" * 40) 
```
